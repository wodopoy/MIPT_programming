# Task 1

```python
from random import randint
import random
import string

#filling file with random strings:

height = randint(1, 101)
length = randint(1, 101)
def random_string(length):
    letters = string.printable
    rand_string = ''.join(random.choice(letters) for i in range(length))
    return rand_string

with open("text.txt", "w") as text:
    for i in range(height):
        text.write(random_string(length))
        #text.write(text"\n")
        
#reading the file without seeing spaces (and tabs):

with open("text.txt", "r") as text:
    for line in text:
        i = 0
        while line[i] == ' 'or line[i] == '\t':
            i += 1
        j = 0
        while line[j - 1] == ' ' or line[j - 1] == '\t':
            j -= 1
        if j == 0:
            j = len(line)
        print(line[i:j])
```

# Task 2

```python
def write_array(array, file_name):
    """записывает строки из array в файл file_name"""
    lines = map(lambda x: x + '\n', array)
    with open(str(file_name) + ".txt", "w") as text:
        text.writelines(lines)
    pass
```

```python
array = [
    'first line',
    'SeConD LiNe',
    'ttthhhiiirrrddd llliiinnneee',
    '4th 1!93',
    '5'
]
file_name = "text2"

write_array(array, file_name)

with open("text2.txt", "r") as file:
    for line in file:
        print(line)
```

# Task 3

Вам дана в архиве файловая структура, состоящая из директорий и файлов.

Вам необходимо распаковать этот архив (средствами языка python), и затем найти в данной в файловой структуре все директории, в которых есть хотя бы один файл с расширением “.py”.

Ответом на данную задачу будет являться файл со списком таких директорий, отсортированных в лексикографическом порядке.

```python
import zipfile
import os

archive = zipfile.ZipFile('main.zip', 'r')
archive.extractall('.') 
archive.close() 
```

```python
data = []
array = []
for current_dir, dirs, files in os.walk("main"):           
    data.append([current_dir, files])
for pair in data:
    for each in pair[1]:
        if each[-3:] == '.py' and pair[0]:
            array.append(pair[0])
            break
array.sort()
write_array(array, 'text3')
```
